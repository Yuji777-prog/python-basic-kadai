```python
class Human:
    def __init__(self, name, age):
        # 属性の初期化
        self.name = name
        self.age = age

    def printinfo(self):
        # nameとageを標準出力
        print(f"名前: {self.name}")
        print(f"年齢: {self.age}")


# Humanクラスのインスタンスを作成し、変数に代入
person = Human("侍太郎", 36)

# メソッドを呼び出して情報を表示
person.printinfo()
```

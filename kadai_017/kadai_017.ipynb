```python
# 関数の定義
def crate_content(name, status):
  #与えられた引数から文字列を作成し、結果を戻り値として返す
  output = name + " は " + status

  return output

class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    # 連想配列（辞書）から結果を取り出して表示するメソッド
    def check_adult(self, result_dict):
        status = result_dict.get(self.name, "不明")
        print(crate_content(self.name, status))


# Humanクラスのインスタンスを20人分生成
humans = [
    Human("山田一郎", 25),
    Human("山下弘美", 18),
    Human("河野敏夫", 15),
    Human("山本和也", 31),
    Human("伊藤俊樹", 20),
    Human("上田博", 30),
    Human("尾ノ上保志", 24),
    Human("鴨浩二", 34),
    Human("坂本良治", 23),
    Human("下岡忠司", 32),
    Human("瀬戸直樹", 17),
    Human("曽川涼香", 25),
    Human("野田幸一", 35),
    Human("田川亮", 25),
    Human("津川智", 16),
    Human("戸田尚子", 28),
    Human("夏川史郎", 33),
    Human("根本和美", 19),
    Human("橋本雄三", 36),
    Human("細川信弘", 24),
]

# 連想配列（辞書）に name をキーとして、成人かどうかの結果を登録
result_dict = {}
for h in humans:
    if h.age >= 20:
        result_dict[h.name] = "大人です。"
    else:
        result_dict[h.name] = "大人ではありません。"

# check_adultメソッドを使って、連想配列から全員分の結果を取り出して表示
for h in humans:
    h.check_adult(result_dict)
```
